```python
import numpy as np
from matplotlib import pylab

#P1. Función lineal
x = np.arange(-5,5)
y = x
fig = pylab.figure(1, figsize=(5,5))
pylab.plot(x,y, 'r--o', label='Lineal', lw=2)
pylab.title('Función')
pylab.xlabel('x')
pylab.ylabel('y')
pylab.grid(True)
pylab.legend(loc='upper left')
pylab.show()
```

```python
#P2. Función sigmoide
x = np.arange(-5,5)
y = 1/(1+np.exp(-x))
fig = pylab.figure(1, figsize=(5,5))
pylab.plot(x,y, 'r--o', label='Sigmoide', lw=2)
pylab.title('Función')
pylab.xlabel('x')
pylab.ylabel('y')
pylab.grid(True)
pylab.legend(loc='upper left')
pylab.show()
```

```python
#P3. Función ReLU
x = np.arange(-5,5)
y = np.maximum(0, x)
fig = pylab.figure(1, figsize=(5,5))
pylab.plot(x,y, 'r--o', label='ReLU', lw=2)
pylab.title('Función')
pylab.xlabel('x')
pylab.ylabel('y')
pylab.grid(True)
pylab.legend(loc='upper left')
pylab.show()
```

```python
#P4. Función Paso Unitario
x = np.arange(-5,5)
y = (x>=0).astype(int)
fig = pylab.figure(1, figsize=(5,5))
pylab.step(x,y, 'r--o', label='Paso', lw=2)
pylab.title('Función')
pylab.xlabel('x')
pylab.ylabel('y')
pylab.grid(True)
pylab.legend(loc='upper left')
pylab.show()
```
