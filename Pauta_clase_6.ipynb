```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

```python
# P1
x = sp.Symbol('x')
h = sp.Symbol('h')
f1 = 4*x - 5
f1_h = f1.subs(x, x + h)
deriv_f1 = sp.limit((f1_h - f1) / h, h, 0)
print(f"{deriv_f1}")
```

```python
# P2
f2 = sp.sqrt(2*x - 3)
f2_h = f2.subs(x, x + h)
deriv_f2 = sp.limit((f2_h - f2) / h, h, 0)
print(f"{deriv_f2}")
```

```python
# P3
x = sp.Symbol('x')
f3 = x**5 - x**3 + 3
deriv_f3 = sp.diff(f3, x)
print(f"{deriv_f3}")
```

```python
f3_numeric = sp.lambdify(x, f3, "numpy")
deriv_f3_numeric = sp.lambdify(x, deriv_f3, "numpy")

x_vals = np.linspace(-2, 2, 400)
y_vals_f3 = f3_numeric(x_vals)
y_vals_deriv_f3 = deriv_f3_numeric(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals_f3, label=r'$f(x) = x^5 - x^3 + 3$', color='blue')
plt.plot(x_vals, y_vals_deriv_f3, label=r"$f'(x)$", color='red', linestyle='--')
plt.title("Gráfica de la función y su derivada")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# P4
x = sp.Symbol('x')
f4 = sp.ln((sp.exp(x) - 1) / (sp.exp(x) + 1))
deriv_f4 = sp.diff(f4, x)
print(f"{deriv_f4}")
```

```python
f4_numeric = sp.lambdify(x, f4, "numpy")
deriv_f4_numeric = sp.lambdify(x, deriv_f4, "numpy")
x_vals = np.linspace(-2, 2, 400)
y_vals_f4 = f4_numeric(x_vals)
y_vals_deriv_f4 = deriv_f4_numeric(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals_f4, label=r'$f(x) = \ln\left(\frac{e^x - 1}{e^x + 1}\right)$', color='blue')
plt.plot(x_vals, y_vals_deriv_f4, label=r"$f'(x)$", color='red', linestyle='--')
undefined_x = x_vals[np.isnan(y_vals_f4)]
plt.scatter(undefined_x, np.zeros_like(undefined_x), color='black', label="Valores no definidos", zorder=3)
plt.axhline(0, color='black', linewidth=0.5, linestyle='dotted')
plt.axvline(0, color='black', linewidth=0.5, linestyle='dotted')
plt.title("Gráfica de la función y su derivada")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# P5
x = sp.Symbol('x')
num5 = sp.sin(sp.pi * x)
den5 = sp.ln(x)
deriv_f5 = sp.diff(num5, x) / sp.diff(den5, x)
lim_f5 = sp.limit(deriv_f5, x, 1)
print(f"{lim_f5}")
```
