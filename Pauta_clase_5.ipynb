```python
import numpy as np
import matplotlib.pyplot as plt

n = int(1e+7)
x = np.random.uniform(-1, 1, n)
y = np.random.uniform(-1, 1, n)

inside_circle = x**2 + y**2 <= 1
pi_estimate = 4 * np.sum(inside_circle) / n

plt.figure(figsize=(6,6))
plt.scatter(x[inside_circle], y[inside_circle], color='blue', s=1)
plt.scatter(x[~inside_circle], y[~inside_circle], color='red', s=1)
plt.title(f"Aproximación de π: {pi_estimate:.4f}")
plt.xlabel("X")
plt.ylabel("Y")
plt.gca().set_aspect('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000)

relu = np.maximum(0, x)
sigmoid = 1 / (1 + np.exp(-x))
tanh = np.tanh(x)
lineal = x

plt.figure(figsize=(10, 6))
plt.plot(x, relu, label='ReLU', linestyle='--')
plt.plot(x, sigmoid, label='Sigmoide', linestyle='-.')
plt.plot(x, tanh, label='Tanh', linestyle=':')
plt.plot(x, lineal, label='Lineal', linestyle='-')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.title("Funciones de Activación")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.show()
```

```python
import numpy as np
import math as m
import matplotlib.pyplot as plt

def sin_taylor(x, terms):
    result = np.zeros_like(x)
    for n in range(terms):
        coef = (-1)**n / m.factorial(2*n + 1)
        result += coef * x**(2*n + 1)
    return result

x = np.linspace(-np.pi, np.pi, 100)
sin_real = np.sin(x)

plt.figure(figsize=(10,6))
plt.plot(x, sin_real, label="sin(x)", color='black')
for terms in [2, 4, 6, 10]:
    plt.plot(x, sin_taylor(x, terms), label=f"{terms} términos")
plt.legend()
plt.grid()
plt.title("Aproximación de sin(x) con Series de Taylor")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000
t = np.linspace(0, 1, fs, endpoint=False)
f_sin = 5
signal = np.sin(2 * np.pi * f_sin * t)

fft_signal = np.fft.fft(signal)
freqs = np.fft.fftfreq(len(t), 1/fs)

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(t, signal)
plt.title("Señal en el tiempo")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")

plt.subplot(1,2,2)
plt.stem(freqs[:fs//2], np.abs(fft_signal)[:fs//2])
plt.title("Espectro de Frecuencia")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Magnitud")

plt.tight_layout()
plt.show()
```
